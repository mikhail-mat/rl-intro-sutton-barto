```python
import numpy as np
import pandas as pd
```

```python
np.random.seed(42)
```

```python
n = 10 # number of actions
```

```python
q = np.random.standard_normal(10) # true values for each action
```

```python
q
```

```python
# note that action 6 (index 5) with the value 1.579.. is the optimal action
```

```python
Q_greedy = np.zeros(10)
Q_eps_greedy = np.zeros(10)
Q_ucb = np.zeros(10)
ucb_terms = np.zeros(10)
c = 2
```

```python
Q_greedy
```

```python
k_greedy = np.zeros(10)
k_eps_greedy = np.zeros(10)
k_ucb = np.zeros(10)
```

```python
def give_reward(action_idx):
  noise = np.random.normal()
  reward = q[action_idx] + noise
  return reward
```

```python
def find_optimal_action_idx(estimates):
  idx_of_highest_estimates = []
  highest = np.max(estimates)
  for i in range(10):
    if estimates[i] == highest:
      idx_of_highest_estimates.append(i)
  return idx_of_highest_estimates
```

```python
def choose_greedy_action():
  idx_of_highest_estimates = find_optimal_action_idx(Q_greedy)
  greedy_action = np.random.choice(idx_of_highest_estimates)
  return greedy_action
```

```python
def choose_eps_greedy_action():
  idx_of_highest_estimates = find_optimal_action_idx(Q_eps_greedy)
  greedy_action = np.random.choice(idx_of_highest_estimates)
  random_action = np.random.randint(10)
  action = np.random.choice([greedy_action, random_action], p=[0.9, 0.1])
  return action
```

```python
def choose_ucb_action():
  unvisited = np.where(k_ucb == 0)[0]
  if len(unvisited) > 0:
      return np.random.choice(unvisited)
  estimates = np.add(Q_ucb, ucb_terms)
  idx_of_highest_estimates = find_optimal_action_idx(estimates)
  ucb_action = np.random.choice(idx_of_highest_estimates)
  return ucb_action
```

```python
t = 0 # timestep

greedy_total_reward = 0
eps_greedy_total_reward = 0
ucb_total_reward = 0

for i in range(10000):
  t += 1

  # greedy
  action = choose_greedy_action()
  reward = give_reward(action)
  greedy_total_reward += reward

  k_greedy[action] += 1
  Q_greedy[action] += (1/k_greedy[action])*(reward - Q_greedy[action])

  # epsilon-greedy
  action = choose_eps_greedy_action()
  reward = give_reward(action)
  eps_greedy_total_reward += reward

  k_eps_greedy[action] += 1
  Q_eps_greedy[action] += (1/k_eps_greedy[action])*(reward - Q_eps_greedy[action])

  # UCB
  action = choose_ucb_action()
  reward = give_reward(action)
  ucb_total_reward += reward

  k_ucb[action] += 1
  Q_ucb[action] += (1/k_ucb[action])*(reward - Q_ucb[action])

  for j in range(10):
    if k_ucb[j] != 0:
      ucb_terms[j] = c*(np.log(t) / k_ucb[j])**0.5
```
